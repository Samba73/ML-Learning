```python
# Step 1: Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Step 2: Prepare the data
# House sizes (square feet)
X = np.array([1000, 1500, 2000]).reshape(-1, 1)
# House prices (dollars)
y = np.array([200000, 250000, 300000])

# Step 3: Create and train the model
model = LinearRegression()
model.fit(X, y)

# Step 4: Get the coefficients
a = model.coef_[0]
b = model.intercept_

# Print the coefficients
print(f"Coefficient (a): {a}")
print(f"Intercept (b): {b}")

# Step 5: Make a prediction
# Predict the price of a house with size 1200 square feet
new_size = np.array([[1200]])
predicted_price = model.predict(new_size)
print(f"Predicted price for a 1200 square feet house: {predicted_price[0]}")

# Step 6: Plot the data and the regression line
plt.scatter(X, y, color='blue')  # Original data points
plt.plot(X, model.predict(X), color='red', linewidth=2)  # Regression line
plt.xlabel('Size (square feet)')
plt.ylabel('Price (dollars)')
plt.title('House Size vs. Price')
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 2, 3, 4, 5])

# Different alpha values
alphas = [0.1, 1.0, 10.0]

# Plotting the effect of different alphas
plt.figure(figsize=(10, 6))
for alpha in alphas:
    model = Ridge(alpha=alpha)
    model.fit(X, y)
    plt.plot(X, model.predict(X), label=f'alpha={alpha}')
    
plt.scatter(X, y, color='black')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Effect of different alpha values in Ridge Regression')
plt.legend()
plt.show()
```
